## Import Module


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Open driver and maximize the windows
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.maximize_window()
# Set the URL and open it
url1 = "https://www.sociolla.com/6322-acne-pimple-patch?sort=-created_at"
driver.get(url1)
```

```python
from selenium.common.exceptions import NoSuchElementException
import time

link = []

while True:
    # Cari elemen <a> di dalam `div.product__info` yang memiliki `href`
    produk_links = driver.find_elements(By.CSS_SELECTOR, 'div.product__info a')

    # Tambahkan link dari setiap produk ke list `link`
    for i in produk_links:
        href = i.get_attribute('href')
        if href:  # Pastikan href tidak None
            link.append(href)
        print(i.text)  # Cetak informasi produk untuk memverifikasi data yang diambil

    # Coba cari tombol "Next Page" dan cek apakah bisa diklik
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, 'a.pagination-navigation-next')
        
        # Cek apakah tombol "Next" nonaktif atau memiliki atribut yang menandakan tidak bisa diklik
        if 'disabled' in next_button.get_attribute('class') or next_button.get_attribute('aria-disabled') == 'true':
            print("Halaman terakhir sudah dicapai.")
            break

        # Jika tombol bisa diklik, klik tombol "Next"
        next_button.click()
        time.sleep(2)  # Tambahkan waktu tunggu agar halaman dimuat sebelum mencari produk
    except NoSuchElementException:
        # Jika tombol "Next Page" tidak ditemukan, keluar dari loop
        print("Halaman terakhir sudah dicapai.")
        break

# Tampilkan semua link produk yang berhasil diambil
print("Link produk yang ditemukan:")
for url2 in link:
    print(url2)
```

```python
import re
# Fungsi untuk mengekstrak semua frasa setelah kata 'untuk'
def extract_all_uses(description):
    matches = re.findall(r'untuk (.+?)(\.|,|$)', description.lower())  # Cari semua frasa setelah 'untuk'
    uses = [match[0].strip() for match in matches]  # Ambil frasa pertama dari setiap hasil dan hilangkan spasi
    return ', '.join(uses) if uses else 'Other'  # Gabungkan frasa dengan koma jika ada, atau beri label 'Other'

# Tambahkan kolom baru 'uses' berdasarkan deskripsi
data_acne['uses'] = data_acne['description'].apply(extract_all_uses)

# Tampilkan DataFrame
data_acne
```

```python
data_acne['content_profile'] = data_acne[['category', 'sub-category', 'ingredients', 'uses']].agg(' '.join, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk

# Unduh daftar stop words bahasa Indonesia dari nltk
nltk.download('stopwords')
from nltk.corpus import stopwords


# Langkah 2: TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words=stopwords.words('indonesian'))
tfidf_matrix = tfidf_vectorizer.fit_transform(data_acne['content_profile'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Fungsi untuk rekomendasi berdasarkan input user
def recommend_products(selected_subcategory, selected_composition, selected_use_for, df=data_acne, tfidf_matrix=tfidf_matrix):
    # Gabungkan input user menjadi "profil user"
    user_profile = f"{selected_subcategory} {selected_composition} {selected_use_for}"
    user_vector = tfidf_vectorizer.transform([user_profile])
    
    # Hitung kesamaan dengan semua produk
    similarity_scores = cosine_similarity(user_vector, tfidf_matrix).flatten()
    
    # Urutkan skor kesamaan
    recommended_indices = similarity_scores.argsort()[-10:][::-1]  # Top 5 rekomendasi
    
    # Ambil detail produk yang direkomendasikan
    recommendations = data_acne.iloc[recommended_indices][['name', 'brand', 'price', 'rating','sub-category']]
    return recommendations

```

```python
# Contoh input user
selected_subcategory = "Acne Cream"
selected_composition = "Niacinamide"
selected_use_for = "mencerahkan"

# Hasil rekomendasi
recommendations = recommend_products(selected_subcategory, selected_composition, selected_use_for)
print("Rekomendasi Produk:")
print(recommendations)
```
