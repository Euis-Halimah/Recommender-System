```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
```

## Touch-ups Sunscreen

```python
# Open driver and maximize the windows
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.maximize_window()
# Set the URL and open it
touchup = "https://www.sociolla.com/6334-touch-ups-sunscreen?sort=-created_at"
driver.get(touchup)
```

```python
from selenium.common.exceptions import NoSuchElementException
import time

link = set()

while True:
    # Cari elemen <a> di dalam `div.product__info` yang memiliki `href`
    produk_links = driver.find_elements(By.CSS_SELECTOR, 'div.product__info a')

    # Saat menambahkan link:
    for i in produk_links:
        href = i.get_attribute('href')
        if href:  # Pastikan href tidak None
             link.add(href)  # Tambahkan ke set
    print(i.text)

    # Coba cari tombol "Next Page" dan cek apakah bisa diklik
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, 'a.pagination-navigation-next')
        
        # Cek apakah tombol "Next" nonaktif atau memiliki atribut yang menandakan tidak bisa diklik
        if 'disabled' in next_button.get_attribute('class') or next_button.get_attribute('aria-disabled') == 'true':
            print("Halaman terakhir sudah dicapai.")
            break

        # Jika tombol bisa diklik, klik tombol "Next"
        next_button.click()
        time.sleep(2)  # Tambahkan waktu tunggu agar halaman dimuat sebelum mencari produk
    except NoSuchElementException:
        # Jika tombol "Next Page" tidak ditemukan, keluar dari loop
        print("Halaman terakhir sudah dicapai.")
        break

# Konversi ke list jika perlu:
link = list(link)

# Tampilkan semua link produk yang berhasil diambil
print("Link produk yang ditemukan:")
for url in link:
    print(url)
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

link
product_data = []

for url in link:
    driver.get(url)  # Buka halaman produk
    time.sleep(2)  # Tunggu beberapa detik agar halaman sepenuhnya dimuat

    try:
        # Ambil data atribut dari halaman detail produk
        product_name = driver.find_element(By.CSS_SELECTOR, 'h1.title').text
        product_brand = driver.find_element(By.CSS_SELECTOR, 'p.brand').text
        product_price = driver.find_element(By.CSS_SELECTOR, 'li.pricing.hasdiscount').text
        product_description = driver.find_element(By.ID, 'product-descriptions').text

        # Cek elemen rating, jika tidak ada masukkan 0
        try:
            rating = driver.find_element(By.CSS_SELECTOR, 'p.rate-value').text
        except Exception:
            rating = "0"

        # Klik tombol "INGREDIENTS" jika ada
        try:
            ingredients_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="mainbody"]/div/div/div[1]/div[2]/div/div/div/div/div/div/div[2]/form/div[2]/ul/li[3]/a'))
            )
            # Scroll ke elemen untuk memastikan terlihat
            driver.execute_script("arguments[0].scrollIntoView(true);", ingredients_button)
            driver.execute_script("arguments[0].click();", ingredients_button)

            # Tunggu elemen untuk informasi bahan muncul
            product_ingredients = WebDriverWait(driver, 10).until(
                EC.visibility_of_element_located((By.XPATH, '//*[@id="ingredients"]/div/p'))
            ).text
        except Exception as e:
            product_ingredients = "Tidak ada informasi bahan"
            print(f"Error saat mengambil informasi bahan: {e}")

        # Simpan data ke dalam list
        product_data.append({
            "name": product_name,
            "brand": product_brand,
            "price": product_price,
            "description": product_description,
            "ingredients": product_ingredients,
            "rating": rating,
        })

    except Exception as e:
        print(f"Terjadi kesalahan saat mengambil data dari {url}: {e}")

# Tambahkan kolom "sub-category" dan "category" ke setiap produk
for product in product_data:
    product["sub-category"] = "Touch ups Sunscreen"
    product["category"] = "Sun Care"

# Tampilkan data produk yang diambil
for product in product_data:
    print(product)
```

```python
import pandas as pd

touchup = pd.DataFrame(product_data)
```

```python
# Mengganti karakter '\n' dengan spasi di kolom 'description' dan 'ingredients'
touchup['description'] = touchup['description'].str.replace('\n', ' ', regex=False)
touchup['ingredients'] = touchup['ingredients'].str.replace('\n', ' ', regex=False)
# Mengambil harga setelah '\n' jika ada, dengan pengecekan untuk menghindari error
touchup['price'] = touchup['price'].apply(
    lambda x: x.split('\n')[1] if '\n' in x and len(x.split('\n')) > 1 else x
)
touchup
```

```python
touchup.to_csv("Touch_Ups_Sunscreen.csv", sep=',', index=False, encoding='utf-8')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Gantilah ini dengan daftar URL yang ingin kamu scrape
link

driver = webdriver.Chrome()
driver.set_page_load_timeout(60) 

product_data = []

for url in link:
    driver.get(url)  # Buka halaman produk
    time.sleep(2)  # Tunggu beberapa detik agar halaman sepenuhnya dimuat

    try:
        # Ambil data atribut dari halaman detail produk
        product_name = driver.find_element(By.CSS_SELECTOR, 'h1.title').text
        product_brand = driver.find_element(By.CSS_SELECTOR, 'p.brand').text
        product_price = driver.find_element(By.CSS_SELECTOR, 'li.pricing.hasdiscount').text
        product_description = driver.find_element(By.ID, 'product-descriptions').text

        # Cek elemen rating, jika tidak ada masukkan 0
        try:
            rating = driver.find_element(By.CSS_SELECTOR, 'p.rate-value').text
        except Exception:
            rating = "0"

        # Blok try untuk mengambil informasi bahan
        try:
            # Klik tombol "INGREDIENTS" jika ada
            ingredients_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="mainbody"]/div/div/div[1]/div[2]/div/div/div/div/div/div/div[2]/form/div[2]/ul/li[3]/a'))
            )
            driver.execute_script("arguments[0].scrollIntoView(true);", ingredients_button)
            driver.execute_script("arguments[0].click();", ingredients_button)

            # Tunggu agar konten muncul
            WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.ID, "ingredients"))
            )

            # Coba klik tombol "See More" jika ada
            try:
                see_more_button = WebDriverWait(driver, 3).until(
                    EC.element_to_be_clickable((By.XPATH, '//*[@id="ingredients"]/a')) 
                )
                driver.execute_script("arguments[0].click();", see_more_button)  # Klik tombol
                # Tunggu konten tambahan muncul setelah klik
                WebDriverWait(driver, 5).until(
                    EC.presence_of_element_located((By.XPATH, '//*[@id="ingredients"]/div/p'))
                )
            except Exception:
                pass  # Jika tidak ada tombol, lanjutkan saja

            # Ambil teks bahan setelah memastikan semua konten tampil
            product_ingredients = WebDriverWait(driver, 10).until(
                EC.visibility_of_element_located((By.XPATH, '//*[@id="ingredients"]/div/p'))
            ).text

        except Exception as e:
            product_ingredients = "Tidak ada informasi bahan"
            print(f"Error saat mengambil informasi bahan: {e}")

        # Simpan data ke dalam list
        product_data.append({
            "name": product_name,
            "brand": product_brand,
            "price": product_price,
            "description": product_description,
            "ingredients": product_ingredients,
            "rating": rating,
        })

    except Exception as e:
        print(f"Terjadi kesalahan saat mengambil data dari {url}: {e}")

# Tambahkan kolom "sub-category" dan "category" ke setiap produk
for product in product_data:
    product["sub-category"] = "Face Serum"
    product["category"] = "Treatment"

# Tampilkan data produk yang diambil
for product in product_data:
    print(product)

# Tutup driver setelah selesai
driver.quit()
```
